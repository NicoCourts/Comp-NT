# Project Euler problem 29 -- Distinct Powers
Consider all integer combinations of $a^b$ for $2\le a \le 5$ and $2 \le b \le 5$:

$$2^2=4, 2^3=8, 2^4=16, 2^5=32$$
$$3^2=9, 3^3=27, 3^4=81, 3^5=243$$
$$4^2=16, 4^3=64, 4^4=256, 4^5=1024$$
$$5^2=25, 5^3=125, 5^4=625, 5^5=3125$$
If they are then placed in numerical order, with any repeats removed, we get the following sequence of 15 distinct terms:

$$4, 8, 9, 16, 25, 27, 32, 64, 81, 125, 243, 256, 625, 1024, 3125$$

How many distinct terms are in the sequence generated by $a^b$ for $2 \le a \le 100$ and $2 \le b \le 100$?


---
## Ideas
- This problem is essentially equivalent to counting the number of unique representations certain numbers have as powers: 
$$a^b=c^d$$

```python
def genPrimes(n):
    composites = dict()
    for i in range(2,n):
        if i not in composites:
            yield i
            composites[i*i] = set()
            composites[i*i].add(i)
        else:
            for j in composites[i]:
                nextMultiple = i + j
                if nextMultiple not in composites:
                    composites[nextMultiple] = set()
                composites[nextMultiple].add(j)
            del composites[i]
```

```python
pg = genPrimes(100)
primes = set(p for p in pg)
```

```python
from functools import reduce
import operator
# create a set of possible primes
cols = []
for p in primes:
    cols.append([p])
for i in range(len(primes)+1):
    for lst in cols:
        prod = reduce(operator.mul, lst, 1)
        for p in primes:
            if prod*p <= 100:
                cols.append(lst.append(p))
```
